```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 20})
```

```python
def link_performance(x,t0,alpha,k,beta):
    return t0*(1 + alpha*(x/k)**beta)

beta = 4
alpha = 0.15
k = 0.5
t0 = 1

x = np.linspace(0,1,100)
y  = link_performance(x,t0,alpha,k,beta)


fig,ax = plt.subplots(figsize = (9,5),dpi = 300)
ax.plot(x,1/y,lw = 3)
ax.set_title("1 / link performance function",fontsize = 16)
ax.set_ylim(0,1.2)
ax.set_ylabel("1 / Travel Time")
ax.set_xlabel("Vehicle Flow")

plt.grid()
```
